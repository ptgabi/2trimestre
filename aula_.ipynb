```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the counts of each unique response for the question
faculdade_counts = dados['você pensa em fazer uma faculdade?'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x=faculdade_counts.index, y=faculdade_counts.values)
plt.title('Distribuição de Respostas: ')
plt.xlabel('Resposta')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()Você pensa em fazer uma faculdade?
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the counts of each unique response for the question
curso_counts = dados['pretende fazer qual curso?'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6)) # Increased figure size for potentially more unique courses
sns.barplot(x=curso_counts.index, y=curso_counts.values)
plt.title('Distribuição de Respostas: Pretende fazer qual curso?')
plt.xlabel('Curso')
plt.ylabel('Contagem')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

## Resumo do histograma
* A maioria dos participantes pratica atividade física entre 0 e 1 vez por semana. Isso pode ser visto pela alta frequência de respostas próximas a esses valores.
* Poucos participantes praticam atividade física de maneira mais intensa (de 4 a 7 vezes por semana), sugerindo que uma prática regular de exercícios ainda não é comum entre os respondentes.
* Essa distribuição é muito concentrada em valores baixos, o que pode indicar uma tendência de baixa frequência de atividade física na amostra.
